First I'm going to create the data simulation process.

```julia
using Plots
using Statistics 

σ = 0.2 
α = [0.8 1.0 1.2]
t_max = 200
T_a = zeros(100)
means = zeros(3)

for j in 1:100
    x = zeros(201)
    x[1] = 1.0
    for i in 2:t_max 
        ϵ = randn()
        x[i] = α[2]*x[i-1] + σ*ϵ
        if x[i]<0 && T_a[j]==0
            T_a[j] = i
        end 
        if i==t_max && T_a[j]==0
            T_a[j]= t_max
        end   
    end
end
means[2] = mean(T_a)
histogram(T_a.-1, bins=0:10:200)
```

**Exercise 8(a) - Introductory Examples**

```julia
using LinearAlgebra
function fixedpointmap(f, f_prime, x_0, tolerance=1E-7, maxiter=1000)
    # setup the algorithm
    x_old = x_0
    normdiff = Inf
    iter = 1
    while normdiff > tolerance && iter <= maxiter
        x_new = x_old-f(x_old)/f_prime(x_old) # use the passed in map
        normdiff = norm(x_new - x_old)
        x_old = x_new
        iter = iter + 1
    end
    return (value = x_old, normdiff=normdiff, iter=iter) # A named tuple
end

# Test with the function f(x)=(x-1)^3
f(x)=(x-1)^3
fprime(x)=3*(x-1)^2

# Another possibility is a linear function such as x+k which should never converge.
sol = fixedpointmap(f, fprime, 0.0) # don't need to pass
println("Fixed point = $(sol.value), and |x^new - x^old| = $(sol.normdiff) in $(sol.iter)"*
        " iterations")
```

```julia
# Another possibility is a linear function such as x^a +k which should never converge.
h(x) = x^2 + 3
hprime(x) = 2x
sol = fixedpointmap(h, hprime, 0.0) # don't need to pass
println("Fixed point = $(sol.value), and |x^new - x^old| = $(sol.normdiff) in $(sol.iter)"*
        " iterations")
# Effectively never converges
```

```julia
function factorial2(n)
    x = 1
    for i = 1:n 
        x = x*(n-i+1)
    end
    return(value = x)
end
sol = factorial2(6)
factorial2(6) == factorial(6)
```

```julia
function linapprox(f, range, x)
    # evaluates the piecewise linear interpolant of f at x,
    # on the interval [a, b], with n evenly spaced grid points.
    pt = range[Int(((x-range[1])÷ step(range))+1)]
    w, z = pt, pt +step(range)
    return f(w) + (x-z)*(f(z) - f(w)) / (w-z)
end
```

```julia
f(v)=cos(v)

range_aprox=-1.0:0.5:2.0
range_complete=-1.0:0.002:2.0
interpol=zeros(length(range_complete))
for i in eachindex(interpol)
    interpol[i]=linapprox(f,range_aprox,range_complete[i])
end

plt = plot(xlim = (-1,2), ylim = (0,1))
plot(range_complete,f.(range_complete), label="True Function")
plot!(range_complete,interpol, label="Interpolation")
```
